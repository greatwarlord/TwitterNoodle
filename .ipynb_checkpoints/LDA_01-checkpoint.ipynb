```python
from tweet_feed import Feed
from basic_cleaner import BasicCleaner
import data_object
file_path = "../DataCollection/191107-16_38_54--191107-16_38_59" 
sentiment_range = [float(-1), float(1)]

from nltk.tokenize import word_tokenize

def get_data_list():
    feed = Feed()
    queue_stream = feed.disk_get_tweet_queue(file_path)
    data_objects = [data_object.get_dataobj_converted(tweet) for tweet in queue_stream]
    for obj in data_objects: BasicCleaner.autocleaner(obj,sentiment_range, False)
    return [x.text for x in data_objects]



    
```

```python
def tokenize(_list):
    new_list = []
    for item in _list:
       new_list.append(word_tokenize(item))
    return new_list
    
l = get_data_list()#[:5]
for x in l:
    print(f"{x} \n")


tokens = [section for section in tokenize(l)]
```

```python

from gensim import corpora
dictionary = corpora.Dictionary(tokens)
corpus = [dictionary.doc2bow(text) for text in tokens]

for x in dictionary:
    print(dictionary.get(x))


import gensim
#NUM_TOPICS = int(len(l) / 2)
NUM_TOPICS = 5

def do_modelling():
    ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=10)
    #ldamodel.save('model5.gensim')
    topics = ldamodel.print_topics(num_words=7)
    for topic in topics:
        print(topic)
        
#do_modelling()
        
        
# import pyLDAvis.gensim

# def do_visualisation():
#     lda = gensim.models.ldamodel.LdaModel.load('model5.gensim')
    
#     lda_display = pyLDAvis.gensim.prepare(lda, corpus, dictionary, sort_topics=False)
#     pyLDAvis.display(lda_display)
    
# def do_visualisation():
#     print("staring visualisation func \n")
#     lda = gensim.models.ldamodel.LdaModel.load('model5.gensim')
#     pyLDAvis.enable_notebook()
#     vis = pyLDAvis.gensim.prepare(lda, corpus, dictionary, sort_topics=False)
#     print(type(vis))

# do_visualisation()
    
```
